```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Creación de un array de numpy con valores de inversión en I+D
investment = np.array([2, 12, 5, 20, 8, 15, 25, 18, 7, 22, 10, 17, 30, 14, 19, 6, 24, 27, 13, 21, 9, 11, 28, 16, 26, 23])

# Creación de un array de numpy con el número correspondiente de patentes generadas
number_patents = np.array([1, 8, 4, 15, 6, 10, 20, 12, 5, 18, 7, 11, 25, 9, 14, 3, 16, 22, 8, 13, 6, 7, 23, 10, 21, 19])

# Redimensionamiento de la matriz de inversión para que sea compatible con el modelo de regresión
x = investment.reshape(-1, 1)

# Asignación de la variable dependiente (número de patentes)
y = number_patents

# Division de los datos en conjuntos de entrenamiento (80%) y prueba (20%)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Creacion de un modelo de regresión lineal
model = LinearRegression()

# Entrenamiento del modelo con los datos de entrenamiento
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

# Calculo del coeficiente de determinación R^2, que indica la precisión del modelo
r2 = model.score(x_test, y_test)

# Obtención del coeficiente (pendiente) de la regresión lineal
coefficient = model.coef_[0]

# Obtención del término independiente (intercepto) de la regresión lineal
intercept = model.intercept_

print("Linear Regression")
print("R^2 Score:", r2)

print("Coefficient:", coefficient)
print("Intercept:", intercept )

# Impresión de los resultados predichos de patentes en función de la inversión
print("R&D Investment VS Number of Patents")
for investent, predictedPatents in zip(y_test, y_pred):
    print("Invesment :", investent, "Predicted Patents:", predictedPatents)

# Visualización de la relación entre inversión y número de patentes utilizando un diagrama de dispersión
plt.scatter(x, y, color='purple', label='Investment')

# Trazado de la línea de regresión con los datos de prueba y las predicciones
plt.plot(x_test, y_pred, color='blue', linewidth=2, label='Predicted Patents')

# Configuración de las etiquetas y el título del gráfico
plt.xlabel('Investment k$', color="blue")
plt.ylabel('Final Patents', color="purple")
plt.title('Technology: Relationship between R&D investment and number of patents', color="green")
plt.legend()
plt.show()


```
